```python
import pandas as pd
nba=pd.read_excel('nba.xlsx')
nba.head(16)
```

```python
def jugador(posicion):
    global pivot
    global base
    global escolta
    global alero
    global alapivot
    if posicion == 'P':
        pivot=nba.loc[nba['Rebotes']== nba['Rebotes'].max()]
        print(pivot)
    if posicion == 'A':
        alero=nba.loc[nba['Triples']== nba['Triples'].max()]
        print(alero)
    if posicion == 'B':
        base=nba.loc[nba['Asistencias']== nba['Asistencias'].max()]
        print(base)
    if posicion == 'AP':
        alapivot=nba.loc[nba['Robos']== nba['Robos'].max()]
        print(alapivot)
    if posicion == 'E':
        escolta=nba.loc[nba['Puntos']== nba['Puntos'].max()]
        print(escolta)
```

```python
jugador('P')
jugador('A')
jugador('B')
jugador('E')
jugador('AP')
```

```python
class nba:        
    def mejoresjugadores(self):
        mejores = pd.concat([pivot,alero,base,alapivot,escolta])
        mejores.groupby('Nombre')['Salario'].sum().plot(kind='bar')
```

```python
basket = nba()
basket.mejoresjugadores()
```
